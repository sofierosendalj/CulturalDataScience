## Music sales challenge

Write a script that:

1. Combines the tables of best-selling physical singles and best-selling digital singles on the Wikipedia page "List_of_best-selling_singles"
2. Adds a column which marks whether each row is from the list of physical singles or digital singles
3. Outputs the artist and single name for the year you were born. If there is no entry for that year, take the closest year after you were born.
4. Outputs the artist and single name for the year you were 15 years old.

### Combining and adding a coloumn

```python
# Starter code...

%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_singles")

# Combining the best singles from rawdata variable
df_physical = rawdata[0]
df_digital = rawdata[3]

# Adds a column which marks whether each row is from the list of physical singles or digital singles
df_physical['Type'] = 'Physical Single'
df_digital['Type'] = 'Digital Single'

best_singles = pd.concat([df_physical, df_digital]) # giving a list of dataframes and concat them
best_singles
```

### Output of the artist and single name for the closest year I was born

```python
# Saving birth year in variable 
birth_year = 2000 

# Going into coloumn with release year, and output the ones equal to or after my birth day
year_singles = best_singles.loc[best_singles['Released'] >= birth_year] 
year_singles
```

```python
# Sorting the dataframe by the released year to get it in order 
closest_year = year_singles.sort_values(by=['Released'])
# Taking the first item in the sorted list, since this must be the single closest to the year 2000
closest_year = closest_year.iloc[0]
# Only outputting the artist name and single for the year 
print(f"Artist: {closest_year['Artist']}, Single: {closest_year['Single']}")
```

### Output of the artist and single name for the year I was 15 years old

```python
songs_2015 = best_singles[best_singles['Released'] == 2015] # acces the release year row and check for instances = 2015
songs_2015
```

## Space challenge

1. Make a single dataframe that combines the space missions from the 1950's to the 2020's
2. Write a script that returns the year with the most launches
3. Write a script that returns the most common month for launches
4. Write a script that ranks the months from most launches to fewest launches


### Space missions from 1950's to 2020's

```python
# Starter code...

#%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/Timeline_of_Solar_System_exploration")

# Combining space missions from index 0 to 8 from the rawdata, since this marks 1950's to 2020's
df_space_missions = pd.concat(rawdata[0:8])
df_space_missions
```

### Returning the year with the most launches

```python
# Creating if statement to check the coloumn 'Launch date' in dataframe

if 'Launch date' in df_space_missions.columns:
    # Count launches per year:
    # Using value_counts() to count the occurrences of each year, which represents the number of launches
    launch_counts = df_space_missions['Launch date'].value_counts()
    
    # Get the year with the maximum launches:
    # Use idxmax() to find the year with the highest count and max() to get the number of launches
    most_launches_year = launch_counts.idxmax()
    max_launches = launch_counts.max()

print(f"The year with the most launches is {most_launches_year} with {max_launches} launches.")
```

### Returning the most common month for launches

```python

# Check if there's a date column (assuming the column is named 'Launch Date')
if 'Launch date' in df_space_missions.columns:
    # Convert 'Launch Date' to datetime format
    df_space_missions['Launch date'] = pd.to_datetime(df_space_missions['Launch date'])
    
    # Extract the month from 'Launch Date'
    # .dt.month_name() makes is write the name of the month instead of ex "10" for October
    df_space_missions['Month Name'] = df_space_missions['Launch date'].dt.month_name() 
      
    # Count launches per month
    month_counts = df_space_missions['Month Name'].value_counts()

    most_common_month = month_counts.idxmax()
    max_launches = month_counts.max()

    print(f"The most common month for launches is {most_common_month} with {max_launches} launches.")
    
```

### Write a script that ranks the months from most launches to fewest launches


```python
# usring .sort_values to sort the dataframe
df_months_sorted = month_counts.sort_values(ascending = False)
df_months_sorted
```
