# Tabulating and cross tabulating data

### Frequency/count tables

```python

data = {'Dinner':["lasagna",  "pasta",  "pasta",  "noodles",  "greek salad",   "tacos",   "soup", "noodles",
  "cury", "grilled cheese"],
       'Weekday':["monday", "tuesday", "wednesday", "thursday", "friday",  "saturday",  "sunday", "monday", "tuesday", "wednesday"]}

df = pd.DataFrame(data, columns=['Dinner','Weekday'])

df
```

```python
# You could check the frequency for a meal
pd.crosstab(index = df["Dinner"], columns = "count")
```

```python
# If we want it displayed on all the weekdays, we can get a frequency table 
# - we change the  argument in coloums from "count" to weekdays
pd.crosstab(index=df["Dinner"], columns=df["Weekday"],margins=True)
```

### Converting frequency/count tables to a table of proportions

```python
tabs = pd.crosstab(index=df["Dinner"], columns=df["Weekday"],margins=True)

tabs.columns = list(tabs.columns)[0:-1] + ['rowtotals']
tabs.index = list(tabs.index)[0:-1] + ['coltotals']

tabs
```

```python
tabs/tabs.loc['coltotals'] # when listed like this, we see that pasta and noodles each took up 20% of the dinners
```

# Extracting a subset from a dataframe

```python
age = [17, 19, 21, 37, 18, 19, 47, 18, 19]
score = [12, 10, 11, 15, 16, 14, 25, 21, 29]
rt = [3.552, 1.624, 6.431, 7.132, 2.925, 4.662, 3.634, 3.635, 5.234]
group = ["test", "test", "test", "test", "test", "control", "control", "control", "control"]

df = pd.DataFrame(
    {'age': age,
     'score': score,
     'rt': rt,
     'group': group
    })

df
```

```python
# We can take the scores by age

df_old = df[df['age']> 21] 
df_old
```

```python
# We can add in more parameters
old_and_slow_control = df[(df['age'] > 21) & (df['rt'] > 3) & (df['group'] == 'control')]
old_and_slow_control
```

# Reshaping a dataframe (format)

```python

df_drugs = pd.read_csv("https://raw.githubusercontent.com/ethanweed/pythonbook/main/Data/drugs1.csv")
```

```python
df_drugs
```

```python
# this is a wide-form dataset; there is an within-subject factor, that can be handles with pandas.melt()
df_long = pd.melt(df_drugs, id_vars=[ 'id', 'gender'])
df_long.head(13)
```

```python
# A more complex dataset could look like this
df = pd.read_csv("https://raw.githubusercontent.com/ethanweed/pythonbook/main/Data/drugs.csv")
df.head()
```

```python
# Here we might want to reshape the dataset from wide to long format
df_long = pd.wide_to_long(df, stubnames = ['WMC','RT'], 
                          i=['id','gender'], 
                          j='drug', 
                          sep = '_', 
                          suffix = '.+')
df_long.head(10)
```

```python
df_long.head(4) # Multiindex
```

```python
# We throw out the old index information and reset the index
df_long = df_long.reset_index() 
# This will now show our wide format data in a long format
df_long.head(15)
```

```python
# We can also do it the other way around, so the long data format will be shown in wide format
df_wide = pd.pivot(df_long, index=['id'], columns='drug', values=['gender', 'WMC', 'RT'])
df_wide
```

# Loops

```python
# While loop 
sweet_enough = 3
num_sugar_spoons = 0
while num_sugar_spoons < sweet_enough:
    num_sugar_spoons = num_sugar_spoons + 1
    print('I have added', num_sugar_spoons, 'spoons of sugar')
```

```python
# For loop
eggs = ['egg', 'egg', 'egg']
for egg in eggs:
    print('I cracked an egg!')
```

```python
# For in loop

ingredients = ['eggs', 'flour', 'sugar', 'cinnamon', 'salt']
for item in ingredients:
    print('I added', item)
```

```python
words = ['eggs', 'flour', 'sugar', 'cinnamon', 'salt']

for word in words:
    print('There are', len(word), 'letters in', word.upper())
```

```python
# Conditional statement
day = 'Wednesday'

if day == 'Wednesday':
    print('Bike Class!')
```

```python
# If-statement in for loop
week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

for day in week:
    if day == 'Wednesday':
        print('Today is', day + '.', 'You have a bike class!')
    else:
        print('Today is', day + '.', 'You can leave at the normal time today.')
```

```python
# You can add 'or' to create more statements in loop
week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

for day in week:
    if day == 'Wednesday':
        print('Today is', day + '.', 'You have a bike class!')
    if day == 'Saturday' or day == 'Sunday':
            print('Today is', day + '.', 'You can sleep late!')
    else:
        print('Today is', day + '.', 'You can leave at the normal time today.')
```

```python
# If AND logic - we can keep adding new conditions with "if"
numbers = [4, 2, 54, 823452, 324, 2, 4435, 4, 9070878072634, 3421, 4345]
for num in numbers:
    if len(str(num)) > 1:
        if num % 2 == 0:
            if int(str(num)[-1]) == 4:
                print(num)
```
